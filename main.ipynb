```python
import pandas as pd 
import matplotlib.pyplot as plt
import os
```

```python

os.chdir(r"C:\Users\ayush\OneDrive\Desktop\nilesh dai assignment\DAI-101-Assignment-1\dataset")
df = pd.read_csv("pokedex.csv")
print(df.head(10))
```

```python
columns_to_analyze = ['HP', 'Attack', 'Defense', 'SP. Atk.', 'SP. Def', 'Speed', 'Total']
# df["HP"].describe()

def univariate_analysis(data,col):
    print("Column Name:",col)
    print(data[col].describe())
    plt.figure(figsize=(8,6))
    plt.hist(data[col],bins = min(20,data[col].max()/5),edgecolor = "blue")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
    print("\n"+"_"*50+"\n")
    
for i in columns_to_analyze:
    univariate_analysis(df,i)
    
```
